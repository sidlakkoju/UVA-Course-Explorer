```python
import numpy as np
import pandas as pd
import pickle


csv_file = "multi_semester_v2.csv"
df = pd.read_csv(csv_file)
df
```

```python
embedding_strings = df['embeddings']
embeddings = np.array([np.array(eval(embedding)) for embedding in embedding_strings])
embeddings.shape
```

```python
# index to data dict

index_to_data_dict = {}

for index, row in df.iterrows():
    index_to_data_dict[index] = {
        'level': row['acad_career_descr'],
        'catalog_number': row['catalog_nbr'],
        'class_number': row['class_nbr'],
        'subject': row['subject_descr'],
        'name': row['descr'],
        'credits': row['units'],
        'description': row['descriptions'],
        'mnemonic': row['subject'],
        'strm': row['strm']}

```

```python
# data to index dict
data_to_index = {}
for index in index_to_data_dict.keys():
    data = index_to_data_dict[index]
    mnemonic_number_tuple = (data['mnemonic'], data['catalog_number'])
    data_to_index[mnemonic_number_tuple] = index
```

```python
# pickling the things
import pickle

# Store embedding matrix in a pickle file
with open('embedding_matrix.pkl', 'wb') as embedding_file:
    pickle.dump(embeddings, embedding_file)

# Store data dictionary in a pickle file
with open('index_to_data_dict.pkl', 'wb') as data_dict_file:
    pickle.dump(index_to_data_dict, data_dict_file)


with open('data_to_index_dict.pkl', 'wb') as index_dict_file:
    pickle.dump(data_to_index, index_dict_file)
```
