```python
import pandas as pd

df = pd.read_csv("multi_semesters_with_embeddings.csv")
df
```

```python
# check where things are getting duplicated

columns_to_check = ['acad_org', 'subject', 'catalog_nbr']
duplicate_rows = df[df.duplicated(subset = columns_to_check, keep=False)]
grouped = duplicate_rows.groupby(columns_to_check)

# Iterate through the groups and print the rows side by side
for _, group in grouped:
    group_subset = group[["acad_org", "subject", "catalog_nbr", "strm"]]
    print(group_subset.to_string(index=False))
print(len(grouped))
```

```python
df_filtered = df.drop_duplicates(subset=['catalog_nbr', 'subject'])
print(len(df_filtered))
```

```python
value_counts = df_filtered['strm'].value_counts()
value_counts
```
