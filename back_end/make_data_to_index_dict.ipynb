```python
import pickle

# first input the existing course_data dict
with open('course_data_dict.pkl', 'rb') as f:
    course_data_dict = pickle.load(f)

data_to_index = {}
for index in course_data_dict.keys():
    data = course_data_dict[index]
    mnemonic_number_tuple = (data['mnemonic'], data['catalog_number'])
    data_to_index[mnemonic_number_tuple] = index
```

```python
with open('data_to_index_dict.pkl', 'wb') as f:
    pickle.dump(data_to_index, f)
```
